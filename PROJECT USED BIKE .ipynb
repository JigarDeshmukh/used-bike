```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#  READ &CLEANING OF DATA AND FORMATTING IT PROPERLY

```python
data=pd.read_csv("usedbikedata")
```

```python
data   #here there is unnamed :0 columns 
```

```python
data.rename(columns={'Unnamed: 0':"Sr_no"},inplace=True)  #renaming done of (unnamed:0) to sr_no
```

```python
data.head()
```

```python
data.set_index('Sr_no',inplace=True) # changing  index no by using set_index
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

# FINDING THE TOTAL COUNT OF  BRAND  | WHICH BRAND HAS MORE DEMAND IN RESALE MARKET


```python
plt.figure(figsize=(10,10))
data['brand'].value_counts().plot(kind='bar')
plt.title("BIKE BRAND")
plt.ylabel("NO of bike")
plt.xlabel("BRAND")
plt.xticks(rotation=90)
plt.show()
```

```python
data.value_counts("brand") #finding which brand has more bike in resale market
```

# TOTAL NUMBER OF BIKE IN LIST WITH BIKE NAME

```python
plt.figure(figsize=(50,50))
data['bike_name'].value_counts().plot(kind="barh") # pandas plot
plt.title("Bike Brand")
plt.ylabel("brand")
plt.xlabel("count")
plt.xticks(rotation=90)
plt.show()
```

```python
data.value_counts("bike_name")   # for quick overview 
```

```python
data.value_counts("bike_name").idxmax()
```
